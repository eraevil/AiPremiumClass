```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np
X,y= load_iris(return_X_y=True) #鸢尾花数据集，花萼长度，花萼宽度，花瓣长度，花瓣宽度
X_filtered= X[:100]
y_filtered= y[:100]
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y_filtered, test_size=0.25 )
 
```

```python
bias=0
theat=np.random.randn(1,4)
#print(theat)
```

```python
# 模型运算
def forward(x, theat, bias):
    z=np.dot(theat,x.T)+bias
    y_hat=1/(1+np.exp(-z))# sigmoid函数激活，load_iris是二分leist数据集

    return y_hat
```

```python
#计算损失

def loss_fun(y,y_hat):
    e=1e-9
    return - y * np.log(y_hat+e)-(1-y)*np.log(1-y_hat+e)

#print(y_hat)
#loss_fun(y_train,y_hat)
```

```python
#计算梯度
def calc_gradient(x,y,y_hat):
    m=x.shape[0]
    # 梯度下降法
    # θ的偏导，可描述函数在某一点的变化率，梯度越小，变化率越小,趋近于0时，就说明参数已经收敛了。
    delta_theat=np.dot((y_hat-y),x)/m
     # bias偏导
    delta_bias=np.mean(y_hat-y)
    return delta_theat,delta_bias
#calc_gradient(X_train,y_train,y_hat)


# 梯度下降训练
lr = 0.06
epochs = 3000

for i in range(epochs):
    y_hat = forward(X_train,theat,bias)
    #计算损失
    loss_val = loss_fun(y_train,y_hat)
    #计算梯度
    delta_theat,delta_bias = calc_gradient(X_train,y_train,y_hat)
    #更新参数
    theat = theat - lr*delta_theat
    bias = bias - lr*delta_bias

    acc=np.mean(np.round(y_hat)==y_train)

    print(f"epoch:",{i},"loss:",{np.mean(loss_val)},"acc:",acc)

print(theat,bias)
```

```python
idx=np.random.randint(0,len(X_test))
x=  X_test[idx]
y=  y_test[idx]
predict=np.round(forward(x,theat,bias))
print(f"y:{y},predict:{predict}")
```
