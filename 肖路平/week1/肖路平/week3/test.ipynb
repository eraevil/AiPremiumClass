```python
import torch
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
X, y = load_iris(return_X_y=True)
X=X[:100]
y=y[:100]
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

#X_train
#y_train

#创建张量数据
tensor_x = torch.tensor(X, dtype=torch.float32)
tensor_y = torch.tensor(y, dtype=torch.float32)

learning_rate = 0.01
epochs = 50000

#模型参数
w=torch.randn(1,4,requires_grad=True) #向量的梯度
b=torch.randn(1,requires_grad=True)

```

```python
for i in range(epochs):
    #前向运算
    z=torch.nn.functional.linear(tensor_x,w,b) #线性运算
    z=torch.sigmoid(z) #结果是张量
    #z.shape
    #损失函数,所有误差和的平均值，结果是张量
    loss=torch.nn.functional.binary_cross_entropy(z.reshape(-1),tensor_y,reduction="mean")
    #z.shape
    #z.reshape(-1)#-1时，将张量 拉平
    #z.squeeze().shape #去处维度为1的维度
    loss
    #计算梯度
    loss.backward()#计算梯度，梯度值会保存在w.grad和b.grad中

    with torch.no_grad():#关闭梯度计算跟踪
        w -= learning_rate*w.grad
        b -= learning_rate*b.grad

        # 清零梯度
        w.grad.zero_()
        b.grad.zero_()

    print("train loss",loss.item())
```
