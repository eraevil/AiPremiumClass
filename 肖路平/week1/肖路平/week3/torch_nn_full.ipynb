```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision.transforms.v2 import ToTensor #转换图像数据为张量
from torchvision.datasets import FashionMNIST
from torch.utils.data import DataLoader #数据加载器
```

```python
LR=1e-3
epochs=20
BATCH_SIZE=128 #批处理大小
```

```python
#加载数据集 train=true 是训练集,false是测试集
train_data = FashionMNIST(root='./fashion_data', train=True, download=True,transform=ToTensor())
test_data = FashionMNIST(root='./fashion_data', train=False, download=True,transform=ToTensor())
```

```python
train_dl=DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True) #shuffle=True 打乱数据
for x,y in train_dl:
    print(x.shape,y.shape) 
```

```python
#模型创建，结构串联
#model =nn.Sequential(
#    nn.Linear(784, 64),
#    nn.Sigmoid(),
#    nn.Linear(64, 10) 
#    )

#模型创建，结构串联
model2 =nn.Sequential(
    nn.Linear(784, 128),
    nn.Sigmoid(),
    nn.Linear(128, 10) 
)

model=nn.Sequential(
    nn.Linear(784, 128),
    nn.Sigmoid(),
    nn.Linear(128, 64),
    nn.Sigmoid(),
    nn.Linear(64, 10) 
)

```

```python
#损失函数
loss_fn = nn.CrossEntropyLoss()#交叉熵损失函数
#优化器（模型参数更新）
optimizer = torch.optim.SGD(model.parameters(), lr=LR)#model.parameters()包含所有模型参数，lr是学习率 

```

```python
for epoch in range(epochs):  
  for data ,target in train_dl:
    #前向运算
      output = model(data.reshape(-1,784))
      #计算损失
      loss = loss_fn(output,target)
      #反向传播
      optimizer.zero_grad()#清空所有参数梯度，因为在每次反向传播（loss.backward()）时，PyTorch 默认会将梯度累加到现有的梯度值上，而不是覆盖它们。
      loss.backward() #计算梯度（参数.grad）
      optimizer.step() #更新参数
      
      print(f"epoch:",{epoch},"loss:",{loss.item()})
```

```python
test_dl=DataLoader(test_data,batch_size=BATCH_SIZE,shuffle=True)

correct=0
total=0
with torch.no_grad():
    for data,target in test_dl:        
        output=model(data.reshape(-1,784))
        _,predicted =torch.max(output.data,1)
        total += target.size(0)
        correct += (predicted==target).sum().item()
 
print(f"Accuracy:{correct/total*100}%")
```
