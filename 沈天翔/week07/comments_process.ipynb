```python
import csv
from turtle import st
import jieba
# Comment Star

file_name = 'douban_moves_comments.csv'

moves_comments = []
with open(file_name, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        star = int(row['Star'])
        comment = row['Comment']
        # print(star)
        # print(comment)
        if star in [1, 2, 4, 5]:
            # print(comment)
            words = jieba.lcut(comment)
            moves_comments.append((words, 1 if star > 3 else 0 ))

len(moves_comments)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

comments_len = [len(comment) for comment, star in moves_comments]
plt.hist(comments_len, bins=100)
plt.show()
```

```python
moves_comments1 = [c for c in moves_comments if len(c[0]) in range(10, 80)]
print(len(moves_comments1))
```

```python
comments_len = [len(comment) for comment, star in moves_comments1]
plt.hist(comments_len, bins=100)
plt.show()
```
