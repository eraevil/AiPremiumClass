```python
import fasttext
```

```python
###分词处理
import jieba 

with open('三国演义.txt','r',encoding='utf-8') as f:
    lines=f.read()

with open('sgyy_c.txt','w') as f:
    f.write(' '.join(jieba.cut(lines)))
```

```python
model=fasttext.train_unsupervised('sgyy_c.txt',model='skipgram')

print('文档词汇表',model.words)

print('文档词汇表长度',len(model.words))
```

```python
###获取词向量
print(model.get_word_vector('桃园结义'))
```

```python
##获取近邻词
print(model.get_nearest_neighbors('桃园结义',k=5))
```

```python
##分析词间类比
print(model.get_analogies('桃园结义','关羽','曹操'))
```

```python
word1 = "桃园结义"
word2 = "关羽"
similarity = model.get_word_vector(word1).dot(model.get_word_vector(word2))
print(similarity)

word3 = "刘备"
similarity1 = model.get_word_vector(word1).dot(model.get_word_vector(word3))
print(similarity1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
vector1 = model.get_word_vector(word1).reshape(1, -1)
vector2 = model.get_word_vector(word2).reshape(1, -1)
cosine_sim = cosine_similarity(vector1, vector2)[0][0]
print(f"{word1} 和 {word2} 的余弦相似度: {cosine_sim}")
```

```python
##模型保存
model.save_model('sgyy.bin')

###加载模型
model=fasttext.load_model('sgyy.bin')
```
